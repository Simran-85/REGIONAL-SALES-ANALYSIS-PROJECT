```python
#Data cleaning and wrangling

df = df_sales.merge(
    df_customers,
    how='left',
    left_on='Customer Name Index',
    right_on='Customer Index'
)  
```

```python
df.head(5)
```

```python
#merge with products

df = df.merge(
    df_products,
    how='left',
    left_on='Product Description Index',
    right_on='Index')
```

```python
df.head(5)
```

```python
#merge with regions
df = df.merge(
    df_regions,
    how='left',
    left_on='Delivery Region Index',
    right_on='id')
```

```python
df.head(5)
```

```python
#merge with state regions
df = df.merge(
    df_state_reg[["State Code","Region"]],
    how='left',
    left_on='state_code',
    right_on='State Code'
)
```

```python
df.head(5)
```

```python
#merge with budget
df = df.merge(
    df_budgets,
    how='left',
    on='Product Name'
)
```

```python
#customer index
#index
#id
#state code

#clean up redundant columns

cols_to_drop = ['customer Index', 'Index', 'id','State Code']
df = df.drop(columns = cols_to_drop,errors='ignore') 

df.head(5)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Convert 'order_date' to datetime if not already done
df['order_date'] = pd.to_datetime(df['order_date'], format='%Y-%m-%d')

# Step 2: Extract month name (e.g., Jan, Feb, etc.)
df['month'] = df['order_date'].dt.month
df['month_name'] = df['order_date'].dt.strftime('%b')  # 'Jan', 'Feb', etc.

# Step 3: Group by calendar month across all years and sum revenue
monthly_seasonality = df.groupby(['month', 'month_name'])['revenue'].sum().reset_index()

# Step 4: Sort by numeric month (to ensure JanDec order)
monthly_seasonality = monthly_seasonality.sort_values('month')

# Step 5: Plot the seasonality line chart
plt.figure(figsize=(10, 6))
plt.plot(monthly_seasonality['month_name'], monthly_seasonality['revenue'], marker='o', color='orange', linewidth=2)
plt.title('Overall Monthly Sales Seasonality (All Years Combined)', fontsize=16)
plt.xlabel('Calendar Month')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Group by product_name and calculate revenue and order count
product_stats = df.groupby('product_name').agg(
    total_revenue=('revenue', 'sum'),
    order_count=('order_number', 'count')  # Assuming each row is a unique order line
).reset_index()

# Step 2: Sort by total revenue and select top 10 products
top_10_products = product_stats.sort_values(by='total_revenue', ascending=False).head(10)

# Step 3: Melt the DataFrame for grouped bar plot
top_10_melted = top_10_products.melt(
    id_vars='product_name',
    value_vars=['total_revenue', 'order_count'],
    var_name='Metric',
    value_name='Value'
)

# Step 4: Plot using Seaborn
plt.figure(figsize=(14, 6))
sns.barplot(
    data=top_10_melted,
    x='product_name',
    y='Value',
    hue='Metric',
    palette='Set2'
)

# Step 5: Customizing the plot
plt.title('Top 10 Products by Revenue and Order Count', fontsize=16)
plt.xlabel('Product Name')
plt.ylabel('Value')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Metric')
plt.tight_layout()
plt.show()
```

```python
# 1 Compute average profit margin percentage for each channel
channel_margin = (
    df.groupby('channel')['profit_margin_pct']  # Group by sales channel
      .mean()                                  # Calculate mean profit margin %
      .sort_values(ascending=False)            # Sort channels from highest to lowest margin
)

# 2 Set the figure size for clarity
plt.figure(figsize=(6, 4))

# 3 Plot a bar chart of average profit margin by channel
ax = sns.barplot(
    x=channel_margin.index,    # X-axis: channel names
    y=channel_margin.values,   # Y-axis: average profit margin values
    palette='coolwarm'         # Color palette for bars
)

# 4 Add chart title and axis labels
plt.title('Average Profit Margin by Channel')  # Main title
plt.xlabel('Sales Channel')                    # X-axis label
plt.ylabel('Avg Profit Margin (%)')             # Y-axis label

# 5 Annotate each bar with its exact margin percentage
for i, v in enumerate(channel_margin.values):
    ax.text(
        i,                  # X position (bar index)
        v + 0.5,            # Y position (bar height + small offset)
        f"{v:.2f}%",        # Text label showing percentage with two decimals
        ha='center',        # Center-align the text horizontally
        fontweight='bold'   # Bold font for readability
    )

# 6 Adjust layout to prevent clipping and display the plot

plt.tight_layout()
plt.show()

```

```python
# Aggregate total revenue and unique order count per state
state_rev = df.groupby('state_name').agg(
    revenue=('revenue', 'sum'),          # Sum up revenue per state
    orders=('order_number', 'nunique')   # Count unique orders per state
).sort_values('revenue', ascending=False).head(10)  # Keep top 10 by revenue

# Plot 1: Top 10 states by revenue (scaled to millions)
plt.figure(figsize=(15, 4))
sns.barplot(
    x=state_rev.index,                   # X-axis: state names
    y=state_rev['revenue'] / 1e6,        # Y-axis: revenue in millions
    palette='coolwarm'                   # Color palette
)
plt.title('Top 10 States by Revenue')     # Chart title
plt.xlabel('State')                       # X-axis label
plt.ylabel('Total Revenue (Million USD)') # Y-axis label
plt.tight_layout()                        # Adjust layout
plt.show()                                # Display the plot

# Plot 2: Top 10 states by number of orders
plt.figure(figsize=(15, 4))
sns.barplot(
    x=state_rev.index,                   # X-axis: state names
    y=state_rev['orders'],               # Y-axis: order counts
    palette='coolwarm'                   # Color palette
)
plt.title('Top 10 States by Number of Orders')  # Chart title
plt.xlabel('State')                              # X-axis label
plt.ylabel('Order Count')                        # Y-axis label
plt.tight_layout()                               # Adjust layout
plt.show()                              
```

```python
# Aggregate metrics per customer
cust_summary = df.groupby('customer_name').agg(
    total_revenue=('revenue', 'sum'),
    total_profit=('profit', 'sum'),
    avg_margin=('profit_margin_pct', 'mean'),
    orders=('order_number', 'nunique')
)

# Convert revenue to millions
cust_summary['total_revenue_m'] = cust_summary['total_revenue'] / 1e6

plt.figure(figsize=(7, 5))

# Bubble chart with revenue in millions
sns.scatterplot(
    data=cust_summary,
    x='total_revenue_m',        # <-- use revenue in millions
    y='avg_margin',
    size='orders',
    sizes=(20, 200),
    alpha=0.7
)

plt.title('Customer Segmentation: Revenue vs. Profit Margin')
plt.xlabel('Total Revenue (Million USD)')  # <-- updated label
plt.ylabel('Avg Profit Margin (%)')

plt.tight_layout()
plt.show()
```

```python
# List numeric columns to include in the correlation calculation
num_cols = ['order_quantity','unit_price', 'revenue', 'cost', 'profit']

# Calculate the correlation matrix for these numeric features
corr = df[num_cols].corr()

# Set the figure size for clarity
plt.figure(figsize=(6,4))

# Plot the heatmap with annotations and a viridis colormap
sns.heatmap(
    corr,           # Data: correlation matrix
    annot=True,     # Display the correlation coefficients on the heatmap
    fmt=".2f",      # Format numbers to two decimal places
    cmap='viridis'  # Color palette for the heatmap
)

# Add title for context
plt.title('Correlation Matrix')

# Adjust layout to prevent clipping
plt.tight_layout()

# Display the heatmap
plt.show()
```
